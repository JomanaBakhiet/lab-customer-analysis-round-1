```python
import pandas as pd  
import numpy as np
```

```python
file1 = pd.read_csv("files_for_lab/csv_files/file1.csv")
file2 = pd.read_csv("files_for_lab/csv_files/file2.csv")
file3 = pd.read_csv("files_for_lab/csv_files/file3.csv")
```

```python
#Standardize header names.
```

```python
cols1 = []
for column in list(file1.columns):
    column = column.lower().replace(' ', '_')
    cols1.append(column)
file1.columns = cols1
```

```python
cols2 = []
for column in list(file2.columns):
    column = column.lower().replace(' ', '_')
    cols2.append(column)
file2.columns = cols2
```

```python
cols3 = []
for column in list(file3.columns):
    column = column.lower().replace(' ', '_')
    cols3.append(column)
file3.columns = cols3
```

```python
display(file1)
display(file2)
display(file3)
```

```python
file2 = file2 [['customer','st', 'gender', 'education', 'income', 'vehicle_class', 'monthly_premium_auto', 'customer_lifetime_value', 'policy_type', 'number_of_open_complaints', 'total_claim_amount']]
display(file2)
```

```python
file3 = file3 [['customer','state', 'gender', 'education', 'income', 'vehicle_class', 'monthly_premium_auto', 'customer_lifetime_value', 'policy_type', 'number_of_open_complaints', 'total_claim_amount']]
display(file3)
```

```python
# Correct the values in the column 'customer lifetime value'. They are given as a percent, so divide them by 100 and change dtype to numerical type. (only 2 of the files need this step.)
```

```python
file1['customer_lifetime_value']
```

```python
file1['customer_lifetime_value'] = file1['customer_lifetime_value'].str.replace('%', '').astype(float)
```

```python
file1['customer_lifetime_value'] = file1['customer_lifetime_value']/100
```

```python
display(file1)
```

```python
file2['customer_lifetime_value']
```

```python
file2['customer_lifetime_value'] = file2['customer_lifetime_value'].str.replace('%', '').astype(float)
```

```python
file2['customer_lifetime_value'] = file2['customer_lifetime_value']/100
```

```python
display(file2)
```
